```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
labels = ["Java", "Python", "PHP", "JavaScript", "Cp", "Cpp"]
slices = [1258,2556,2222,1598,2556,5664]
explode =[0,0.1,0,0,0,0]
plt.pie(slices,labels=labels,shadow=True,explode=explode,autopct='%1.1f%%',startangle=340,wedgeprops={'edgecolor':'black'})
plt.title(' my pie chart')
plt.tight_layout()
plt.show()
```

# **  stack plot **

```python
minutes = [1,2,3,4,5,6,7,8,9]

players1= [1,2,3,4,4,6,7,5,6]
players2= [1,5,2,1,1,2,8,7,2]
players3 =[1,2,5,4,6,3,7,2,5]
labels= ['players1','players2','players3']
colors=['r','b','g']

plt.stackplot(minutes,players1,players2,players3,labels=labels,colors=colors)
plt.legend(loc='upper left')
plt.title('my stack  plot')
```
