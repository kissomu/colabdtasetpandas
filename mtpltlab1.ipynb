```python
import matplotlib.pyplot as plt
```

```python
bat_x = [5,6,7,8,9,11]
run_y = [222,389,459,585,655,755]
run_z = [255,777,881,981,1001,1100]
run_w = [261,581,881,981,1111,1151]
plt.plot(bat_x,run_y,color='b',linestyle='--',marker='^',linewidth=3)
plt.plot(bat_x,run_z,color='b',linestyle='-',marker='.')
plt.plot(bat_x,run_w,color='m',linestyle='-',marker='.')
plt.legend(['run made by india','run made by aus','run made by SA'])
plt.title('cricket bat number and run that made by that bat number')
plt.xlabel('bat number')
plt.ylabel('run made by that bat number')
plt.tight_layout()
plt.grid()



plt.show()
```
